```python
all_movies_genres['Processed Genre'] = all_movies_genres['Genres'].apply(lambda x: re.findall('\w+', x))
```

```python
all_genres = []
for index, row in all_movies_genres.iterrows():
#     print(row['Genres'])
    all_genres += row['Processed Genre']
    
```
