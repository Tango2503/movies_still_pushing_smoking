```python
def f(x):
    d = {}
    d['total_smoke'] = x['has_smoke'].sum()
    d['total_movies'] = x['Title'].count()
    d['total_incidents'] = x['Total incidents'].sum()
    d['total_descriptors'] = x['MPAA descriptor'].sum()
    return pd.Series(d, index = ['total_smoke','total_movies','total_incidents','total_descriptors'])

byyear_smoke = all_movies.groupby(['Year']).apply(f).reset_index()
byyear_smoke['Percentage'] = byyear_smoke['total_smoke']/byyear_smoke['total_movies']
byyear_smoke
```

```python
def generate_col_names(attribute, n):
    all_columns = []
    for i in range(1,(n+1)):
        attribute_no = attribute + ' ' + str(i)
        all_columns.append(attribute_no)
    return all_columns
```

```python
def combine_columns(row, attribute):
    all_attributes = []
    for i in range(1,12):
        attribute_no = attribute + ' ' + str(i)
#         print(actor_no)
#         print(row[actor_no])
#         if row[attribute_no] == row[attribute_no]:
        all_attributes.append(row[attribute_no])
#         else:
#             continue
    return all_attributes
```

```python
all_movies['All actors'] = all_movies[generate_col_names('Actor',11)].apply(combine_columns, args = ['Actor'],axis = 1)
all_movies['All ages'] = all_movies[generate_col_names('Apparent age',11)].apply(combine_columns, args = ['Apparent age'],axis = 1)
all_movies['All status'] = all_movies[generate_col_names('Status',11)].apply(combine_columns, args = ['Status'],axis = 1)
all_movies['All products'] = all_movies[generate_col_names('Tobacco product',11)].apply(combine_columns, args = ['Tobacco product'],axis = 1)
```

```python
all_movies_unrolled = pd.read_excel('All movies unrolled.xlsx')
```

```python
all_smoking_movies_actors = all_movies_unrolled.loc[all_movies_unrolled['actor_name'].isnull() == False]
```

```python
all_smoking_movies_actors.columns
```

```python
def f1(x):
    d = {}
    d['# Actors in Movie'] = x['actor_name'].count()
    d['Total Incidents'] = x['Total incidents'].sum()
    return pd.Series(d, index = ['# Actors in Movie','Total Incidents'])

actors_per_movie = all_smoking_movies_actors.groupby(['Year','Title','MPAA rating']).apply(f1).reset_index()
actors_per_movie['Incidents Per Actor'] = actors_per_movie['Total Incidents']/actors_per_movie['# Actors in Movie']
incidents_per_actor = actors_per_movie.groupby(['Year','MPAA rating'])['Incidents Per Actor'].mean().reset_index()
incidents_per_actor.columns = ['Year','MPAA rating', 'Mean Incidents Per Actor']
median_incidents_per_actor = actors_per_movie.groupby(['Year','MPAA rating'])['Incidents Per Actor'].median().reset_index()
median_incidents_per_actor.columns = ['Year','MPAA rating', 'Median Incidents Per Actor']
incidents_per_actor = pd.merge(incidents_per_actor, median_incidents_per_actor, how = 'inner', on = ['Year','MPAA rating'])
```

```python
incidents_per_actor.plot('Year',['Mean Incidents Per Actor', 'Median Incidents Per Actor'],kind = 'line')
incidents_per_actor.to_excel('Incidents Per Actor.xlsx',index = False)
data_2019 = actors_per_movie.loc[actors_per_movie['Year'] == 2019]
data_2019.sort_values(by = 'Incidents Per Actor')
```
